```python
def augment_image(image):
    image = Image.fromarray(np.uint8(image))
    image = ImageEnhance.Brightness(image).enhance(random.uniform(0.8,1.2))
    image = ImageEnhance.Contrast(image).enhance(random.uniform(0.8,1.2))
    image = ImageEnhance.Sharpness(image).enhance(random.uniform(0.8,1.2))
    image = np.array(image)/255.0
    return image
```

```python
import numpy as np
from PIL import Image, ImageEnhance

def augment_image(image, *args, **kwargs):
    image = Image.fromarray(np.uint8(image))
    for i in range(len(args)):
        if i == 0:
            image = ImageEnhance.Brightness(image).enhance(args[i])
        elif i == 1:
            image = ImageEnhance.Contrast(image).enhance(args[i])
        elif i == 2:
            image = ImageEnhance.Sharpness(image).enhance(args[i])
    for key, value in kwargs.items():
        if key == 'brightness':
            image = ImageEnhance.Brightness(image).enhance(value)
        elif key == 'contrast':
            image = ImageEnhance.Contrast(image).enhance(value)
        elif key == 'sharpness':
            image = ImageEnhance.Sharpness(image).enhance(value)
    image = np.array(image)/255.0
    return image

# Load your image here
image = np.array(Image.open('/content/gdrive/MyDrive/brain_tumor/archive(2)/Testing/glioma/Te-gl_0038.jpg'))

# Augment the image with different parameter values
augmented_image_1 = augment_image(image, 0.8, 1.0, 1.0)
augmented_image_2 = augment_image(image, 1.2, 1.0, 1.0)
augmented_image_3 = augment_image(image, 1.0, 0.8, 1.0)
augmented_image_4 = augment_image(image, 1.0, 1.0, 0.8)
augmented_image_5 = augment_image(image, brightness=0.8)
augmented_image_6 = augment_image(image, contrast=0.8)
augmented_image_7 = augment_image(image, sharpness=0.8)

# Display the augmented images
import matplotlib.pyplot as plt
fig, axs = plt.subplots(1, 8, figsize=(20, 5))
axs[0].imshow(image)
axs[0].set_title('Original Image')
axs[1].imshow(augmented_image_1)
axs[1].set_title('Augmented Image 1')
axs[2].imshow(augmented_image_2)
axs[2].set_title('Augmented Image 2')
axs[3].imshow(augmented_image_3)
axs[3].set_title('Augmented Image 3')
axs[4].imshow(augmented_image_4)
axs[4].set_title('Augmented Image 4')
axs[5].imshow(augmented_image_5)
axs[5].set_title('Augmented Image 5')
axs[6].imshow(augmented_image_6)
axs[6].set_title('Augmented Image 6')
axs[7].imshow(augmented_image_7)
axs[7].set_title('Augmented Image 7')
plt.show()
```

```python
IMAGE_SIZE = 128


def open_images(paths):

    images = []
    for path in paths:
        image = load_img(path, target_size=(IMAGE_SIZE, IMAGE_SIZE))
        image = augment_image(image)
        images.append(image)
    return np.array(images)


images = open_images(train_paths[10:20])
labels = train_labels[10:20]
fig = plt.figure(figsize=(12, 6))
for x in range(1, 9):
    fig.add_subplot(2, 4, x)
    plt.axis('off')
    plt.title(labels[x])
    plt.imshow(images[x])
plt.rcParams.update({'font.size': 12})
plt.show()
```

```python
unique_labels = os.listdir(train_dir)

def encode_label(labels):
    encoded = []
    for x in labels:
        encoded.append(unique_labels.index(x))
    return np.array(encoded)

def decode_label(labels):
    decoded = []
    for x in labels:
        decoded.append(unique_labels[x])
    return np.array(decoded)

def datagen(paths, labels, batch_size=12, epochs=1):
    for _ in range(epochs):
        for x in range(0, len(paths), batch_size):
            batch_paths = paths[x:x+batch_size]
            batch_images = open_images(batch_paths)
            batch_labels = labels[x:x+batch_size]
            batch_labels = encode_label(batch_labels)
            yield batch_images, batch_labels
```

```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Flatten, Dropout, Dense

base_model = VGG16(input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3), include_top=False, weights='imagenet')
base_model.trainable = False

inputs = Input(shape=(IMAGE_SIZE, IMAGE_SIZE, 3))
x = base_model(inputs)
x = Flatten()(x)
x = Dropout(0.3)(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.2)(x)
outputs = Dense(len(unique_labels), activation='softmax')(x)

model = Model(inputs=inputs, outputs=outputs)
```

```python
keras.utils.plot_model(model, show_shapes=True)
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001),
             loss='sparse_categorical_crossentropy',
             metrics=['sparse_categorical_accuracy'])
```

```python
batch_size = 20
steps = int(len(train_paths)/batch_size)
epochs = 80
history = model.fit(datagen(train_paths, train_labels, batch_size=20, epochs=80),epochs=80, steps_per_epoch=steps)
```
